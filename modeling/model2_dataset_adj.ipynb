```python
import pandas as pd
#행렬 연산 패키지
import numpy as np

#데이터 시각화
import matplotlib
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('model2_dataset.xlsx')
dataset
```

```python
a = pd.DataFrame(dataset['insu_risk_expense'].groupby(dataset['customer_id']).sum())
a
```

```python
dataset.drop(['risk','insu_risk_expense'],axis=1,inplace= True)
dataset = dataset.drop_duplicates()
dataset
```

```python
b = pd.merge(dataset,a,on='customer_id')
```

```python
b
```

```python
f = lambda x : 0 if x >0 else 1
b['risk'] = b['insu_risk_expense'].apply(f)
```

```python
model2_dataset_adj = b
```

```python
model2_dataset_adj.to_excel("model2_dataset_adj.xlsx", index=False, encoding='utf-8')
```
