```python
import pandas as pd
```

```python
insu_risk = pd.read_excel('insu_risk.xlsx')
pd.DataFrame.reset_index(insu_risk, drop=True, inplace=True)
insu_risk.head()
```

```python
request_subset = insu_risk[['customer_id', 'insu_risk_expense']]
request_subset.head()
```

```python
request_subset_id_list = list(request_subset['customer_id'])
request_subset_id_list
```

```python
pre_review = pd.read_excel("C:/Users/PIAI/insu_pre_review_processing.xlsx")
pd.DataFrame.reset_index(pre_review, drop=True, inplace=True)
pre_review.head()
```

```python
pre_review['gender'] = pre_review['gender'].map({1:'M', 2:'F'})
```

```python
customer_id_list = list(pre_review['customer_id'])
len(customer_id_list)
```

```python
len(request_subset)
```

```python
result = 0
for request_subset_id in request_subset_id_list:
    if request_subset_id in customer_id_list:
        result += 1

print(result)
```

```python
pre_review[pre_review['customer_id'] == 'C131064'].index[0]
```

```python
request_subset_id_index_list = []

for request_subset_id in request_subset_id_list:
    try:
        request_subset_id_index = pre_review[pre_review['customer_id'] == request_subset_id].index[0]
        request_subset_id_index_list.append(request_subset_id_index)
    except:
        pass
print(request_subset_id_index_list)
```

```python
len(request_subset_id_index_list)
```

```python
model1_dataset = pre_review.iloc[request_subset_id_index_list, :]
model1_dataset = model1_dataset.loc[:, ['customer_id', 'gender', 'age', 'bmi', 'waist', 'bp_min', 'bp_max']]
pd.DataFrame.reset_index(model1_dataset, drop=True, inplace=True)
model1_dataset.head()
```
